# [dbo].[sp_UpdateDepartmentData]

```sql
/****** Object:  StoredProcedure [dbo].[sp_UpdateDepartmentData]    Script Date: 02-09-2025 21:22:50 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_UpdateDepartmentData]
    @DepartmentId INT,
    @DepartmentName NVARCHAR(100),
    @Location NVARCHAR(100)
AS
BEGIN
    UPDATE Departments
    SET DepartmentName = @DepartmentName,
        Location = @Location
    WHERE DepartmentId = @DepartmentId;
END
GO
```
